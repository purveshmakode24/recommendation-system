```python
import pandas as pd

df = pd.read_csv('csv/books.csv',on_bad_lines='skip',encoding='latin-1',sep=';')
df.head()
```

```python
df.info()
```

```python
# No of duplicate records with same title 

# len(df) - len(df['Book-Title'].unique())


df.duplicated(subset='Book-Title').sum()
```

```python
df1 = df.drop_duplicates(subset='Book-Title')

print(len(df1))
```

```python
df1.duplicated(subset='Book-Title').sum()
```

```python
sample_size = 8000
df8k = df1.sample(n=sample_size, replace=False, random_state=490)

df8k = df8k.reset_index()
df8k = df8k.drop('index',axis=1)
```

```python
df8k.head()
```

```python
df8k[['Book-Title', 'Book-Author', 'Publisher']]
```

```python
def clean_text(text):
    result = str(text).lower()
    return result.replace(" ", "")
```

```python
# df1[['Book-Title', 'Book-Author', 'Publisher']].head()

df8k['Book-Author'] = df8k['Book-Author'].apply(clean_text)

df8k[['Book-Title', 'Book-Author', 'Publisher']].head()
```

```python
df8k['Book-Title'] = df8k['Book-Title'].str.lower()
df8k['Publisher'] = df8k['Publisher'].str.lower()
```

```python
df8k[['Book-Title', 'Book-Author', 'Publisher']]
```

```python
df8k
```

```python
_df2 = df8k.drop(['ISBN','Image-URL-S','Image-URL-M','Image-URL-L','Year-Of-Publication'],axis=1)

_df2
```

```python
df2 = _df2
df2['data'] = df2['Book-Title'].astype('str')+ ' ' + df2['Book-Author'].astype('str') + ' ' + df2['Publisher'].astype('str')

print(df2['data'])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

#test = pd.DataFrame({'data': ['Hello Beautiful Word', 'Hello There']})

vectorizer = CountVectorizer()
vectorized = vectorizer.fit_transform(df2['data'])
```

```python
# xx = {'a': ['hello', 'there'], 'b': [1, None]}

# test = pd.DataFrame(xx)


# test.dropna()
print(vectorizer.get_feature_names_out())
# print(vectorized)
print(vectorized.toarray())
print(vectorized.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity(vectorized)

print(similarities)
print(similarities.shape)
```

```python
# Remapping the simailarties vector back to the previos dataframe

df2 = pd.DataFrame(similarities , columns=df8k['Book-Title'], index=df8k['Book-Title']).reset_index()
df2['Year-Of-Publication'] = df1['Year-Of-Publication']
df2['Image-URL-M'] = df1['Image-URL-M']
df2.head()
```

```python
input_book = 'norby and the oldest dragon'
recommendations = pd.DataFrame(df2.nlargest(11,input_book)[['Book-Title', 'Year-Of-Publication', 'Image-URL-M']])
recommendations = recommendations[recommendations['Book-Title']!=input_book]
recommendations
```
